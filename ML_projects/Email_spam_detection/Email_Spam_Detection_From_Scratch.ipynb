```python
# Ensure text is string, lowercase, and clean
df['text'] = df['text'].astype(str)  # Convert to string
df['text'] = df['text'].apply(lambda x: x.lower())  # Apply lower() properly

df['text'][0]
```

```python
# removing puntuation marks from the text
import string

df['text'] = df['text'].apply(lambda x: x.translate(str.maketrans('', '',string.punctuation)))

df['text'][0]
```

```python
# lemmatization the words of pos tagged

import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

# Function to map NLTK POS tags to WordNet POS tags
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ  # Adjective
    elif tag.startswith('V'):
        return wordnet.VERB  # Verb
    elif tag.startswith('N'):
        return wordnet.NOUN  # Noun
    elif tag.startswith('R'):
        return wordnet.ADV  # Adverb
    else:
        return wordnet.NOUN  # Default to noun


lemmatized_text = []

for sentence in pos_tagged:
  lemmatized_sentence = [lemmatizer.lemmatize(word,get_wordnet_pos(tag)) for word,tag in sentence]
  lemmatized_text.append(lemmatized_sentence)



print("Tokenized:", tokenized_text[0],'\n')
print("POS Tagged:", pos_tagged[0],'\n')
print("Lemmatized:", lemmatized_text[0],'\n')
```

# convert lemmatized words into numerical vectors
this can be done by joining list of words and apply tf-idf

```python
# Convert list of words into a single string per document
documents = [' '.join(words) for words in lemmatized_text]

documents[0]
```
